```python
import os
import time
import numpy as np
from numpy.polynomial.hermite import hermgauss
# import pennylane as qml
# import pennylane.numpy as pnp
import matplotlib.pyplot as plt
from itertools import combinations
import copy
```

```python
n_qubits = 8            # number of qubits (be careful: statevector dim = 2^n_qubits)
n_enc, n_dec = 1, 1     # encoding / decoding layers
scope = [-np.pi, np.pi]
phi_std = 0.7
phi_trues = np.linspace(-np.pi, np.pi, 50)

vqcnni_path = f"VQ-CNNI/{n_qubits}/vqc_{n_enc}_{n_dec}"
vqi_path = f"VQI/{n_qubits}/vqc_{n_enc}_{n_dec}_{phi_std}"
vqcnni_phi_preds = np.load(os.path.join(vqcnni_path, "phi_preds.npy")) 
vqi_phi_preds = np.load(os.path.join(vqi_path, "phi_preds.npy")) 

vqcnni_err = np.angle(np.exp(1j * (vqcnni_phi_preds - phi_trues))) ** 2
vqi_err =  np.angle(np.exp(1j * (vqi_phi_preds - phi_trues))) ** 2
```

```python
import matplotlib.pyplot as plt
import numpy as np

path='./'

plt.rcParams.update({
    "font.size": 10,
    "axes.labelsize": 11,
    "axes.titlesize": 11,
    "legend.fontsize": 9,
    "xtick.labelsize": 9,
    "ytick.labelsize": 9,
    "figure.dpi": 400,
    "savefig.dpi": 800,
    "font.family": "serif"
})

# ===============================================================
# 21 graph
# ===============================================================
fig, (ax1, ax2) = plt.subplots(
    2, 1, figsize=(7, 6), constrained_layout=True
)

# ===============================================================
# (a) Predicted  vs True 
# ===============================================================
ax1.scatter(
    phi_trues, vqi_phi_preds,
    s=40, alpha=0.9, color='#56B4E9', edgecolors='gray', linewidth=0.4, label="VQI"
)

ax1.scatter(
    phi_trues, vqcnni_phi_preds,
    s=40, alpha=0.9, color='#E69F00', edgecolors='gray', linewidth=0.4, label="VQ-CNNI"
)

# 45-degree reference line
lims = [
    min(scope[0], vqi_phi_preds.min()),
    max(scope[1], vqi_phi_preds.max())
]
ax1.plot(
    lims, lims,
    linestyle='--', linewidth=1.2,
    label=r"Ideal $\hat{\phi}=\phi$", zorder=-1
)

ax1.set_xlabel(r"$\phi$")
ax1.set_ylabel(fr"$\hat\phi$")
ax1.legend(loc="upper left")
ax1.grid(True, ls="--", alpha=0.6)
ax1.set_xlim(-np.pi, np.pi)

# first subgraph
ax1.text(
    -0.1, 0.95, "(a)",
    transform=ax1.transAxes,
    fontsize=11, fontweight="bold",
    va="top"
)

# ===============================================================
# (b) MSE() in dB
# ===============================================================
ax2.plot(
    phi_trues,
    10 * np.log10(vqi_err),
    marker="o", markersize=6,
    linewidth=1.4, color="#56B4E9",
    label="VQI"
)

ax2.plot(
    phi_trues,
    10 * np.log10(vqcnni_err),
    marker="o", markersize=6,
    linewidth=1.4, color="#E69F00",
    label="VQ-CNNI"
)

ax2.set_xlabel(r"$\phi$")
# ax2.set_ylabel(r"$\mathrm{\delta}{\phi}_\mathrm{dB}^{2}$")
ax2.set_ylabel(r"$10\log_{10}(\delta\phi^2)$")
ax2.legend(loc="best")
ax2.grid(True, ls="--", alpha=0.6)

# second subgraph
ax2.text(
    -0.1, 0.95, "(b)",
    transform=ax2.transAxes,
    fontsize=11, fontweight="bold",
    va="top"
)

# ===============================================================
# save & show
# ===============================================================
plt.savefig(os.path.join(path, "phi_pred_sigmaphi_panel.png"), dpi=600)
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import inset_axes, mark_inset

vqcnni_loss = np.load(os.path.join(vqcnni_path, "mse_history.npy")) 
vqi_loss = np.load(os.path.join(vqi_path, "mse_history.npy")) 

# ----------------  ----------------
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(vqi_loss, marker='.', label="VQI Loss")
ax.plot(vqcnni_loss, marker='.', label="VQ-CNNI Loss")

ax.set_xlabel("Iteration")
ax.set_ylabel("Loss")
ax.legend()
ax.grid(True)

# ---------------- inset  ----------------
axins = inset_axes(
    ax,
    width="45%",     # inset 
    height="45%",    # inset 
    loc="upper right"
)

x1, x2 = 20, 300
x_zoom = range(x1, x2)
axins.plot(x_zoom, vqi_loss[x1:x2], marker='.', label="VQI")
axins.plot(x_zoom, vqcnni_loss[x1:x2], marker='.', label="VQ-CNNI")

axins.set_xlim(x1, x2)
# axins.set_title("Zoom: Iteration 20300", fontsize=10)
axins.grid(True)

#  y 
ymin = min(
    min(vqcnni_loss[x1:x2]),
    min(vqi_loss[x1:x2])
)
ymax = max(
    max(vqcnni_loss[x1:x2]),
    max(vqi_loss[x1:x2])
)
axins.set_ylim(ymin, ymax)

# inset  legend
axins.legend(fontsize=12)

# ----------------  ----------------
mark_inset(ax, axins, loc1=2, loc2=4, fc="none", ec="0.5")

plt.tight_layout()
plt.savefig("./loss_history.png", dpi=800)
plt.show()
```
